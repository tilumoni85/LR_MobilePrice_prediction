# Model building

```python
# First, let's create feature set(x)/ind var by dropping the target variable/ x in this case is a new df where target var 
# has been dropped
# Then, create the target variable/dependent var (y)


x= df.drop(columns = ['Price'])
y = df['Price']

# battery col is too high so we need to bring it to small range

x['battery']=(x['battery'])/1000
x
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.2)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression,Ridge, Lasso
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
# creating a variable Lr for linear regression
```

```python
Lr= LinearRegression()
Lr.fit(x_train,y_train)
```

```python
y_pred =Lr.predict(x_test)

print( 'R2 score is :',r2_score(y_pred,y_test))
print( 'Mean_absolute_Error is : ',mean_absolute_error(y_pred,y_test))

# It shows 83% of the accuracy rate
```

```python
x= cross_val_score(Lr,x_train,y_train,cv=10,scoring = 'r2')
print('Cross validation score is :',x.mean())
```

```python
# Ridge Regression
# Ridge regression is a type of linear regression that adds a regularization term to the loss function to prevent overfitting. In linear regression, the model tries to minimize the sum of the squared differences between the observed and predicted values. However, when the model has many features or when the features are highly correlated, it can lead to overfitting, where the model performs well on training data but poorly on unseen data.

# Ridge regression addresses this issue by adding a penalty to the size of the coefficients. This penalty term 
# is the sum of the squared values of the coefficients (excluding the intercept).
# By shrinking the coefficients, Ridge regression reduces the model's complexity, leading to better generalization.

# Alpha (α)
# The alpha parameter controls the strength of the regularization. 
# It is also known as the regularization parameter or the penalty term.

# Low alpha value (close to 0): This means less regularization, 
# so Ridge regression behaves more like ordinary linear regression. The model may still overfit if the data is comple
# x or noisy.

# High alpha value: This leads to stronger regularization, shrinking the coefficients more and possibly underfitting 
# the model if alpha is too large.

# In the example you provided (alpha=0.05), the regularization is relatively mild. 
# The model still tries to fit the data closely but with a slight penalty to prevent large coefficients.

```

```python
Rr=Ridge(alpha=0.05)
Rr.fit(x_train,y_train)

y_pred=Rr.predict(x_test)

print("R2_Score is :",r2_score(y_pred,y_test))
print("Mean_Absolute_Error is :",mean_absolute_error(y_pred,y_test))

x=cross_val_score(Rr,x_train,y_train,cv=10,scoring='r2')
print('Cross validation score is:',x.mean())
```

```python
# Key Concepts of Lasso Regression
# L1 Regularization:

# Lasso uses L1 regularization, which adds the absolute values of the coefficients to the cost function. 


# This is similar to the Ridge regression cost function, but with the L1 regularization term instead of the L2 term.

# Alpha (α):

# Like Ridge, the alpha parameter in Lasso controls the strength of the regularization.
# Low alpha: Less regularization, leading to a model similar to ordinary linear regression.
# High alpha: Stronger regularization, resulting in more coefficients being reduced to zero, which simplifies the model.
# When to Use Lasso
# Feature Selection: Lasso is particularly useful when you suspect that many of the features in your dataset are irrelevantor
# redundant. By setting some coefficients to zero, Lasso can help you identify the most important features.

# Interpretability: Lasso's ability to simplify the model by selecting only a subset of the features can make the model easier
# to interpret, especially when dealing with high-dimensional data.
```

```python
Ls=Lasso(alpha=0.001)
Ls.fit(x_train,y_train)

y_pred=Ls.predict(x_test)

print("R2_Score is :",r2_score(y_pred,y_test))
print("Mean_Absolute_Error is :",mean_absolute_error(y_pred,y_test))

x=cross_val_score(Ls,x_train,y_train,cv=10,scoring='r2')
print('Cross validation score is:',x.mean())
```
